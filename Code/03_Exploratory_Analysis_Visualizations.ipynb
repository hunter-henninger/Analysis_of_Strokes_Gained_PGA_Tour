# <u>Exploratory Analysis and Visualizations
-----

## Objective
This notebook is filled with EDA with the goal of understanding trends within the data. With the focus of this study on Strokes Gained and Club Head Speed, various plots are created to illustrate existing patterns and interesting phenomena in professional golf.

-----
#### External Libraries Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

#### Read Cleaned Dataframe

```python
df_pga = pd.read_csv('../Data/Sets/full_pga_data.csv')
```

```python
# create a new dataframe of averages for each player 
df_player = df_pga.groupby('player').mean()
df_player.reset_index(inplace=True)
```

<br><br>
## Visualize Club Head Speed
----

```python
# sort dataframe by Club Head Speed
chs = df_player.sort_values('club_head_speed', ascending = False)
plt.figure(figsize = (15,6))
sns.set(style = 'whitegrid')
sns.barplot(x='player', y='club_head_speed', data=chs, palette = 'Blues_d')
plt.xlabel(xlabel = 'PGA Tour Player', fontsize = 15)
plt.ylabel(ylabel = 'Club Head Speed (mph)', fontsize = 15)
plt.xticks(rotation = 90, fontsize = 10)
plt.ylim(100, 130)
plt.axhline(df_player['club_head_speed'].mean(), ls='--' , c = 'r')
plt.text(43,117, "Average")
plt.title('Average Club Head Speed', fontsize = 22);
```

#### Players to Notice:

- **Bubba Watson** averages the fastest club head speed.
- **Chez Reavie** averages the lowest club head speed.
- **Brandt Snedeker** is far below average.
- **Tiger Woods** lands in the top 6.

<br><br>
## Visualizing Strokes Gained
----

### Total Strokes Gained

```python
# sort dataframe by Strokes Gained: Total
sgt = df_player.sort_values('sg:_total', ascending = False)
plt.figure(figsize = (15,6))
sns.set(style = 'whitegrid')
sns.barplot(x='player', y='sg:_total', data=sgt, palette = 'Blues_d')
plt.xlabel(xlabel = 'PGA Tour Player', fontsize = 15)
plt.ylabel(ylabel = 'Average Total Strokes Gained', fontsize = 15)
plt.axhline(df_player['sg:_total'].mean(), ls='--' , c = 'r')
plt.text(43, 1.1, "Average")
plt.xticks(rotation = 90, fontsize = 10)
plt.title('PGA Strokes Gained: Total', fontsize = 22);
```

#### Players to Notice:

- **Tiger Woods** averages the most Total Strokes Gained by a decent margin.
- **Jordan Spieth**, **Rory McIlroy**, and **Dustin Johnson** have had major success on the PGA Tour and are right at the top.
- **Brandt Snedeker** is just above the average regardless of his Club Head Speed.

### Strokes Gained Off the Tee

```python
# sort dataframe by Strokes Gained: Off the Tee
ott = df_player.sort_values('sg:_off-the-tee', ascending = False)
plt.figure(figsize = (15,6))
sns.set(style = 'whitegrid')
sns.barplot(x='player', y='sg:_off-the-tee', data=ott, palette = 'Blues_d')
plt.xlabel(xlabel = 'PGA Tour Player', fontsize = 15)
plt.ylabel(ylabel = 'Average Strokes Gained: Off the Tee', fontsize = 15)
plt.axhline(df_player['sg:_off-the-tee'].mean(), ls='--' , c = 'r')
plt.text(43, .4, "Average")
plt.xticks(rotation = 90, fontsize = 10)
plt.title('PGA Strokes Gained: Off the Tee', fontsize = 22);
```

#### Players to Notice:

- **Jon Rahm** has recorded the highest average although his Club Head Speed is only slightly above average.
- **Rory McIlroy**, **Bubba Watson**, and **Dustin Johnson** all swing the club fast and gain more strokes Off the Tee than most.
- **Brandt Snedeker** loses strokes Off the Tee on average.

### Strokes Gained Approaching the Green

```python
# sort dataframe by Strokes Gained: Approach the Green
atg = df_player.sort_values('sg:_approach-the-green', ascending = False)
plt.figure(figsize = (15,6))
sns.set(style = 'whitegrid')
sns.barplot(x='player', y='sg:_approach-the-green', data=atg, palette = 'Blues_d')
plt.xlabel(xlabel = 'PGA Tour Player', fontsize = 15)
plt.ylabel(ylabel = 'Average Strokes Gained: Approach the Green', fontsize = 15)
plt.axhline(df_player['sg:_approach-the-green'].mean(), ls='--' , c = 'r')
plt.text(43, .45, "Average")
plt.xticks(rotation = 90, fontsize = 10)
plt.title('PGA Strokes Gained: Approach the Green', fontsize = 22);
```

#### Players to Notice

- **Tiger Woods** is considered by many as the greatest golfer of all time and gains many more strokes as he approaches the green than most professional golfers.
- **Bubba Watson** although swings the club fast, loses strokes as he approaches the green on average.

### Strokes Gained Around the Green

```python
# sort dataframe by Strokes Gained: Around the Green
artg = df_player.sort_values('sg:_around-the-green', ascending = False)
plt.figure(figsize = (15,6))
sns.set(style = 'whitegrid')
sns.barplot(x='player', y='sg:_around-the-green', data=artg, palette = 'Blues_d')
plt.xlabel(xlabel = 'PGA Tour Player', fontsize = 15)
plt.ylabel(ylabel = 'Average Strokes Gained: Around the Green', fontsize = 15)
plt.axhline(df_player['sg:_around-the-green'].mean(), ls='--' , c = 'r')
plt.text(43, .15, "Average")
plt.xticks(rotation = 90, fontsize = 10)
plt.title('PGA Strokes Gained: Around the Green', fontsize = 22);
```

#### Players to Notice

- **Brandt Snedeker** is in the top 6 for Strokes Gained: Around the Green although he loses strokes off the tee.
- **Dustin Johnson** is below average for gaining strokes around the green.
- **Jordan Spieth** has been near the top of each of categories.

### Strokes Gained While Putting

```python
# sort dataframe by Strokes Gained: Putting
putt = df_player.sort_values('sg:_putting', ascending = False)
plt.figure(figsize = (15,6))
sns.set(style = 'whitegrid')
sns.barplot(x='player', y='sg:_putting', data=putt, palette = 'Blues_d')
plt.xlabel(xlabel = 'PGA Tour Player', fontsize = 15)
plt.ylabel(ylabel = 'Average Strokes Gained: Putting', fontsize = 15)
plt.axhline(df_player['sg:_putting'].mean(), ls='--' , c = 'r')
plt.text(43, .25, "Average")
plt.xticks(rotation = 90, fontsize = 10)
plt.title('PGA Strokes Gained: Putting', fontsize = 22);
```

#### Players to Notice

- **Brandt Snedeker** on average, gains the most strokes on the green.
- **Bubba Watson** loses strokes on the green on average.
- Some of the best golfers fall back on the list in the putting category such as **Jordan Spieth**, **Dustin Johnson**, and **Rory McIlroy**.

### Main Takeaways

- The best golfers in the world **gain the most strokes** Off the Tee and Approaching the Green.
- The best golfers in the world **lose the most strokes** Around the Green and Putting.

### Correlations
---

```python
# create dataframe of only strokes gained features
df_sg = df_pga[['sg:_total', 
           'sg:_off-the-tee', 'sg:_approach-the-green', 
           'sg:_around-the-green', 'sg:_putting']]

plt.figure(figsize=(6,2))
sns.set(font_scale = 1)
sns.heatmap(df_sg.corr()[['sg:_total']].sort_values('sg:_total', ascending = False), 
            annot = True,  cmap ='RdBu')
plt.title('Approach the Green is Most Positively Correlated' , fontsize = 11);
```

- Approaching the green affects total strokes gained the most while putting affects total strokes gained the least.

#### Strokes Gained Total

```python
g = sns.set(style = 'whitegrid', font_scale = 1.5)
g = sns.pairplot(df_pga, x_vars = ['sg:_off-the-tee',
                                'sg:_approach-the-green', 'sg:_around-the-green',
                                'sg:_putting'], y_vars = ['sg:_total'], kind='reg', height=5)
g.fig.suptitle("Approach the Green is Most Positively Correlated with Total Strokes Gained", 
               fontsize = 27, y=1.08);
```

- Approach the Green has the steepest curve and most data points in the 2-4 strokes range.
- Putting has the most variance from -4 to 4 strokes.

#### Strokes Gained Off the Tee

```python
g = sns.set(style = 'whitegrid', font_scale = 1.5)
g = sns.pairplot(df_pga, x_vars = ['sg:_total',
                                'sg:_approach-the-green', 'sg:_around-the-green',
                                'sg:_putting'], y_vars = ['sg:_off-the-tee'], 
                 kind='reg', height=5)
g.fig.suptitle("Off the Tee is Negatively Correlated with Around the Green and Putting", 
               fontsize = 27, y=1.08);
```

- Approach the Green has the only upward sloping curve. This demonstrates the notion that players gaining strokes off the tee tend to gain strokes while approaching the green as well.

#### Strokes Gained Approaching the Green

```python
g = sns.set(style = 'whitegrid', font_scale = 1.5)
g = sns.pairplot(df_pga, x_vars = ['sg:_total',
                                'sg:_off-the-tee', 'sg:_around-the-green',
                                'sg:_putting'], y_vars = ['sg:_approach-the-green'], 
                 kind='reg', height=5)
g.fig.suptitle("Approach the Green is Negatively Correlated with Putting", 
               fontsize = 27, y=1.08);
```

- Players that are gaining strokes while approaching the green tend to lose strokes once they are on the green.

#### Strokes Gained Around the Green

```python
g = sns.set(style = 'whitegrid', font_scale = 1.5)
g = sns.pairplot(df_pga, x_vars = ['sg:_total',
                                'sg:_off-the-tee', 'sg:_approach-the-green',
                                'sg:_putting'], y_vars = ['sg:_around-the-green'], kind='reg', height=5)
g.fig.suptitle("Around the Green is Negatively Correlated with Off the Tee", 
               fontsize = 27, y=1.08);
```

- Players that gain strokes around the green tend to lose strokes Off the Tee.

#### Putting

```python
g = sns.set(style = 'whitegrid', font_scale = 1.5)
g = sns.pairplot(df_pga, x_vars = ['sg:_total',
                                'sg:_off-the-tee', 'sg:_approach-the-green',
                                'sg:_around-the-green'], y_vars = ['sg:_putting'], kind='reg', height=5)
g.fig.suptitle("Putting is Negatively Correlated with Each Category", 
               fontsize = 27, y=1.08);
```

- These plots demonstrate the erratic, unpredictable nature of putting. The curves show weak, mostly negative relationships between putting and the other Strokes Gained categories.

## <u>Exploring Features
---

#### Use dataframe for Model 1 to avoid visualizing imputed data

```python
df_pga1 = pd.read_csv('../Data/Sets/model_one.csv')
```

#### Break into strokes gained categories: Off the Tee, Approach the Green, Around the Green, Putting

```python
# Off the Tee
df_OT = df_pga1.loc[:,'sg:_off-the-tee':'club_head_speed']
df_OT['sg:_total'] = df_pga1['sg:_total']

# Approach the Green
df_AG = df_pga1.loc[:,'sg:_approach-the-green':'approaches_from_>_100_yards_(rgh)']
df_AG['sg:_total'] = df_pga1['sg:_total']

# Around the Green
df_ArG = df_pga1.loc[:,'sg:_around-the-green':'scrambling_from_<_10_yards']
df_ArG['sg:_total'] = df_pga1['sg:_total']

# Putting
df_P = df_pga1.loc[:,'sg:_putting':'birdie_or_better_conversion_percentage']
df_P['sg:_total'] = df_pga1['sg:_total']
```

## Correlations and Feature Reduction

### Off the Tee

```python
plt.figure(figsize=(6,4))
sns.set(font_scale = 1)
sns.heatmap(df_OT.corr()[['sg:_total']].sort_values('sg:_total', ascending=False), 
            annot = True,  cmap ='RdBu')
plt.title('Off the Tee Stats with Total Strokes Gained' , fontsize = 12);
```

- All of the correlations are below 0.15 other than Strokes Gained: Off the Tee. 
- Although, it demonstrates weak relationships, driving accuracy, distance, and club head speed all positively influence Total Strokes Gained as expected.

#### Reduce Number of Features

```python
# display off the tee features
df_OT.columns
```

##### No features to combine in the Off the Tee category

### Approach the Green

```python
plt.figure(figsize=(8,6))
sns.set(font_scale = 1)
sns.heatmap(df_AG.corr()[['sg:_total']].sort_values('sg:_total', ascending=False), 
            annot = True,  cmap ='RdBu')
plt.title('Approach the Green Stats with Total Strokes Gained' , fontsize = 13);
```

- This correlation plot shows fairly positive relationships between GIR and going for the green percentages with Total Strokes Gained.
- It appears that Fairway Proximity and Proximity to the Hole cause players to lose strokes.

#### Reduce Number of Features

```python
# display approach the green features
df_AG.columns
```

```python
# combine all approaches between 100 and 200 yards
df_AG['approaches_from_100-200_yards'] = df_AG['approaches_from_100-125_yards'] + \
df_AG['approaches_from_125-150_yards'] + df_AG['approaches_from_150-175_yards'] + \
df_AG['approaches_from_175-200_yards']
```

### Around the Green

```python
plt.figure(figsize=(7,4))
sns.set(font_scale = 1)
sns.heatmap(df_ArG.corr()[['sg:_total']].sort_values('sg:_total', ascending = False), 
            annot = True,  cmap ='RdBu')
plt.title('Around the Green Stats with Total Strokes Gained' , fontsize = 12);
```

- Interestingly, scrambling percentage has a stronger positive relationship with total Strokes Gained than Strokes Gained: Around the Green. This makes sense because the higher a players' scrambling percentage the fewer the strokes.

#### Reduce Number of Features

```python
# display around the green features
df_ArG.columns
```

```python
# combine all scrambling percentages between 10 and 30 yards
df_ArG['scrambling_from_10-30_yards'] = df_ArG['scrambling_from_10-20_yards'] + \
                                        df_ArG['scrambling_from_20-30_yards']
```

### Putting

```python
plt.figure(figsize=(6,7))
sns.set(font_scale = 1)
sns.heatmap(df_P.corr()[['sg:_total']].sort_values('sg:_total', ascending = False), 
            annot = True,  cmap ='RdBu')
plt.title('Putting Stats with Total Strokes Gained' , fontsize = 13);
```

- Interesting to observe the ranking of putting distances. 
    - Anything inside 10 feet is most correlated
    - Putts from 3 or 4 feet are least correlated
- Putting Average has a fairly high negative correlation with Total Strokes Gained implying the importance of putting after hitting the GIR.

#### Reduce Number of Features

```python
# display putting features
df_P.columns
```

```python
# combine all putts between 10 and 25 feet
df_P['putting_from_-_10-25\''] = df_P['putting_from_-_10-15\''] + \
df_P['putting_from_-_15-20\''] + df_P['putting_from_-_20-25\'']
```

- Decided to leave putts from 3, 4, 5, 6, 7 as individual metrics in order to identify a specific distance that affects Strokes Gained the most in the model.
