#### External Libraries Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Read Cleaned Dataframe

```python
df_pga = pd.read_csv('../Data/full_pga_data.csv' , index_col = 'Tournament')
df_pga['date'] = pd.to_datetime(df_pga['date'])
```

```python
# create a new dataframe of averages for each player 
df_player = df_pga.groupby('player').mean()
df_player.reset_index(inplace=True)
```

## Visualize Club Head Speed
----

```python
chs = df_player.sort_values('club_head_speed', ascending = False)
plt.figure(figsize = (15,6))
sns.set(style = 'whitegrid')
sns.barplot(x='player', y='club_head_speed', data=chs, palette = 'Blues_d')
plt.xlabel(xlabel = 'PGA Tour Player', fontsize = 15)
plt.ylabel(ylabel = 'Club Head Speed (mph)', fontsize = 15)
plt.xticks(rotation = 90, fontsize = 10)
plt.ylim(100, 130)
plt.axhline(df_player['club_head_speed'].mean(), ls='--' , c = 'r')
plt.text(43,117, "Average")
plt.title('Average Club Head Speed', fontsize = 22);
```

## Visualize Strokes Gained
----

#### Total Strokes Gained

```python
sgt = df_player.sort_values('sg:_total', ascending = False)
plt.figure(figsize = (15,6))
sns.set(style = 'whitegrid')
sns.barplot(x='player', y='sg:_total', data=sgt, palette = 'Blues_d')
plt.xlabel(xlabel = 'PGA Tour Player', fontsize = 15)
plt.ylabel(ylabel = 'Average Total Strokes Gained', fontsize = 15)
plt.axhline(df_player['sg:_total'].mean(), ls='--' , c = 'r')
plt.text(43, 1.1, "Average")
plt.xticks(rotation = 90, fontsize = 10)
plt.title('PGA Strokes Gained: Total', fontsize = 22);
```

#### Strokes Gained Off the Tee

```python
ott = df_player.sort_values('sg:_off-the-tee', ascending = False)
plt.figure(figsize = (15,6))
sns.set(style = 'whitegrid')
sns.barplot(x='player', y='sg:_off-the-tee', data=ott, palette = 'Blues_d')
plt.xlabel(xlabel = 'PGA Tour Player', fontsize = 15)
plt.ylabel(ylabel = 'Average Strokes Gained: Off the Tee', fontsize = 15)
plt.axhline(df_player['sg:_off-the-tee'].mean(), ls='--' , c = 'r')
plt.text(43, .4, "Average")
plt.xticks(rotation = 90, fontsize = 10)
plt.title('PGA Strokes Gained: Off the Tee', fontsize = 22);
```

#### Strokes Gained Approaching the Green

```python
atg = df_player.sort_values('sg:_approach-the-green', ascending = False)
plt.figure(figsize = (15,6))
sns.set(style = 'whitegrid')
sns.barplot(x='player', y='sg:_approach-the-green', data=atg, palette = 'Blues_d')
plt.xlabel(xlabel = 'PGA Tour Player', fontsize = 15)
plt.ylabel(ylabel = 'Average Strokes Gained: Approach the Green', fontsize = 15)
plt.axhline(df_player['sg:_approach-the-green'].mean(), ls='--' , c = 'r')
plt.text(43, .45, "Average")
plt.xticks(rotation = 90, fontsize = 10)
plt.title('PGA Strokes Gained: Approach the Green', fontsize = 22);
```

#### Strokes Gained Around the Green

```python
artg = df_player.sort_values('sg:_around-the-green', ascending = False)
plt.figure(figsize = (15,6))
sns.set(style = 'whitegrid')
sns.barplot(x='player', y='sg:_around-the-green', data=artg, palette = 'Blues_d')
plt.xlabel(xlabel = 'PGA Tour Player', fontsize = 15)
plt.ylabel(ylabel = 'Average Strokes Gained: Around the Green', fontsize = 15)
plt.axhline(df_player['sg:_around-the-green'].mean(), ls='--' , c = 'r')
plt.text(43, .15, "Average")
plt.xticks(rotation = 90, fontsize = 10)
plt.title('PGA Strokes Gained: Around the Green', fontsize = 22);
```

#### Strokes Gained While Putting

```python
putt = df_player.sort_values('sg:_putting', ascending = False)
plt.figure(figsize = (15,6))
sns.set(style = 'whitegrid')
sns.barplot(x='player', y='sg:_putting', data=putt, palette = 'Blues_d')
plt.xlabel(xlabel = 'PGA Tour Player', fontsize = 15)
plt.ylabel(ylabel = 'Average Strokes Gained: Putting', fontsize = 15)
plt.axhline(df_player['sg:_putting'].mean(), ls='--' , c = 'r')
plt.text(43, .25, "Average")
plt.xticks(rotation = 90, fontsize = 10)
plt.title('PGA Strokes Gained: Putting', fontsize = 22);
```
